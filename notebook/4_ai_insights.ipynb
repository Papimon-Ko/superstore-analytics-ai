#### Import Library

```python
import pandas as pd
from openai import OpenAI
```

```python
client = OpenAI(api_key="YOUR_API_KEY_HERE")  # --- IGNORE ---
```

```python
df = pd.read_csv("data/Superstore_Cleaned.csv")
```

```python
# Sale summary
sales_summary = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
top_region = df.groupby('Region')['Sales'].sum().idxmax()
top_category = sales_summary.sort_values(by='Sales', ascending=False).iloc[0]
```

```python
prompt = f"""
Data Summary:
{sales_summary.to_string(index=False)}

Top Region: {top_region}
Top Category by Profit: {top_category}

Please analyze this sales data and write 3 clear insights in Thai and English.
Focus on trends, performance, and potential business actions.
"""
```

```python
# ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏• AI
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a professional data analyst."},
        {"role": "user", "content": prompt}
    ]
)
```

```python
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•
print("\nüìä AI-Generated Insights:")
print(response.choices[0].message.content)
```
