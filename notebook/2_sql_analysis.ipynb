```python
import pandas as pd
import sqlite3
```

#### Load Data

```python
df = pd.read_csv("data/Superstore_Cleaned.csv")
# Connect to SQLite database 
conn = sqlite3.connect("data/superstore.db")
# Load DataFrame into SQL table
df.to_sql("sales", conn, index=False, if_exists="replace")

print("Data loaded into SQLite database successfully.")
```

#### Sample Query

```python
# 1. Total annual sales
query1 = """
SELECT Order_Year, ROUND(SUM(Sales),2) AS Total_Sales
FROM sales
GROUP BY Order_Year
ORDER BY Order_Year;
"""

# 2. Average sales and profit per product category
query2 = """
SELECT Category, 
       ROUND(SUM(Sales),2) AS Total_Sales,
       ROUND(SUM(Profit),2) AS Total_Profit
FROM sales
GROUP BY Category
ORDER BY Total_Sales DESC;
"""

# 3. Top 10 Most Profitable Customers
query3 = """
SELECT [Customer Name], ROUND(SUM(Profit),2) AS Total_Profit
FROM sales
GROUP BY [Customer Name]
ORDER BY Total_Profit DESC
LIMIT 10;
"""

# 4. Sales by region
query4 = """
SELECT Region, ROUND(SUM(Sales),2) AS Total_Sales
FROM sales
GROUP BY Region
ORDER BY Total_Sales DESC;
"""
```

```python
for i, q in enumerate([query1, query2, query3, query4], start=1):
    print(f"\nðŸ“Š result Query {i}:")
    result = pd.read_sql_query(q, conn)
    print(result)
```
